### Clean columns and convert numerical columns to float type for further computation

```python
def clean_numeric(column):
    def convert(x):
        if pd.isna(x):
            return x
        try:
            x = str(x).strip()
            x = x.replace('m²', '').replace('m', '').replace('phòng', '')
            return float(x)
        except:
            print(f"Không thể chuyển đổi giá trị: {x}")  # Debug
            return pd.NA
    
    return column.apply(convert)

# Xử lý dữ liệu
numeric_columns = ['num_floors', 'num_bed_rooms', 'squared_meter_area', 
                  'length_meter', 'width_meter']

for col in numeric_columns:
    df[col] = clean_numeric(df[col])

string_columns = {
    'district': 'Quận ',
    'ward': 'Phường '
}

for col, prefix in string_columns.items():
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(prefix, '').str.strip()

df.head()
```

### Convert all price to "triệu/m2"

```python
#df.loc[df['price_in_million_per_square_meter'].str.contains(' tỷ/m²'), 'price_in_million_per_square_meter'] = df.loc[df['price_in_million_per_square_meter'].str.contains(' tỷ/m²'), 'price_in_million_per_square_meter'].str.replace(' tỷ/m²','').str.replace('.','').str.replace(',','.').astype(float) * 1000
df.loc[df['price_in_million_per_square_meter'].str.contains(' triệu/m²', na=False), 'price_in_million_per_square_meter'] = df.loc[df['price_in_million_per_square_meter'].str.contains(' triệu/m²', na=False), 'price_in_million_per_square_meter'].str.replace(' triệu/m²','').str.replace(',','.').astype(float)
#df.loc[df['price_in_million_per_square_meter'].str.contains(' đ/m²', na=False), 'price_in_million_per_square_meter'] = df.loc[df['price_in_million_per_square_meter'].str.contains(' đ/m²', na=False), 'price_in_million_per_square_meter'].str.replace(' đ/m²','').str.replace('.','').astype(float) * 0.000001
```

```python
# Tính giá trị NaN cho length_meter hoặc width_meter
df['length_meter'] = df.apply(
    lambda row: row['squared_meter_area'] / row['width_meter'] if pd.isna(row['length_meter']) and not pd.isna(row['width_meter']) and not pd.isna(row['squared_meter_area']) else row['length_meter'], 
    axis=1
)
df['width_meter'] = df.apply(
    lambda row: row['squared_meter_area'] / row['length_meter'] if pd.isna(row['width_meter']) and not pd.isna(row['length_meter']) and not pd.isna(row['squared_meter_area']) else row['width_meter'], 
    axis=1
)

# Xử lý NaN cho num_bed_rooms và num_floors với sai số ±1
df['num_bed_rooms'] = df['num_bed_rooms'].fillna(df['num_bed_rooms'].mean().round() + np.random.choice([-1, 0, 1]))
df['num_floors'] = df['num_floors'].fillna(df['num_floors'].mean().round() + np.random.choice([-1, 0, 1]))
```

### Removing outliers using IQR method

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ boxplot cho các biến số
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_cleaned[['num_floors', 'num_bed_rooms', 'squared_meter_area', 
                            'length_meter', 'width_meter', 'price_in_million_per_square_meter']])
plt.xticks(rotation=45)
plt.title('Boxplot của các biến số')
plt.show()
```

```python
def remove_outlier_IQR(df, series):
    Q1=df[series].quantile(0.25)
    Q3=df[series].quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df[series]<(Q1-1.5*IQR)) | (df[series]>(Q3+1.5*IQR)))]
    return df_final

removed_outliers = df_cleaned
columns_to_remove_outliers = ['num_floors', 'num_bed_rooms', 'squared_meter_area', 'length_meter',
                              'width_meter', 'price_in_million_per_square_meter']
for column in columns_to_remove_outliers:
    removed_outliers = remove_outlier_IQR(removed_outliers, column)

print("The final length of the dataset is", str(len(removed_outliers)), "rows.")
```

### Scale variables

```python
# Dữ liệu sau khi loại bỏ outliers
housing = removed_outliers

# Tách các biến dự đoán và phản hồi (giá)
X = housing.loc[:, housing.columns != 'price_in_million_per_square_meter']
y = housing[['price_in_million_per_square_meter']]

# Các feture cần được scaled
to_be_scaled = ['num_floors', 'num_bed_rooms', 'squared_meter_area', 'length_meter', 'width_meter']

# Khởi tạo scaler
PredictorScaler = StandardScaler()
TargetVarScaler = StandardScaler()

# Chỉ lưu trữ các giá trị scaled
X_scaled = X.copy()
y_scaled = y.copy()

# Chuyển đổi các biến dự đoán
X_scaled[to_be_scaled] = PredictorScaler.fit_transform(X_scaled[to_be_scaled])

# Chuyển đổi biến mục tiêu
y_scaled = TargetVarScaler.fit_transform(y)

# Chuyển đổi thành mảng numpy
X_array = np.array(X_scaled.values).astype("float32")
y_array = np.array(y_scaled).astype("float32")

# Tách dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_array, y_array,
                                                    test_size=0.2,
                                                    random_state=2023)

# Kiểm tra tính hợp lệ của các tập dữ liệu
assert X_train.shape[0] == y_train.shape[0] and X_train.shape[1] == X_test.shape[1] and X_test.shape[0] == y_test.shape[0] and y_train.shape[1] == y_test.shape[1], \
"All train and test sets should have correct dimensions."

# Chuyển đổi đầu vào để phù hợp với CNN/RNN/LSTM
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_cnn = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

```

### Build Model

#### CNN model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

# Xây dựng mô hình CNN
model_cnn = Sequential()

# Layer 1: Tầng Conv1D
model_cnn.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)))

# Layer 2: Tầng MaxPooling1D
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Dropout(0.2))
# Layer 3: Tầng Conv1D
model_cnn.add(Conv1D(32, kernel_size=3, activation='relu'))

# Layer 4: Tầng MaxPooling1D
model_cnn.add(MaxPooling1D(pool_size=2))

model_cnn.add(Dropout(0.2))
# Layer 5: Tầng Flatten để chuyển dữ liệu thành dạng vector
model_cnn.add(Flatten())

# Layer 6: Tầng Dense (fully connected)
model_cnn.add(Dense(64, activation='relu'))

# Layer 7: Tầng Dropout để giảm overfitting
model_cnn.add(Dropout(0.4))

# Output Layer: Đầu ra dự đoán giá nhà
model_cnn.add(Dense(1))  # Dự đoán giá nhà (1 giá trị liên tục)

# Biên dịch mô hình
model_cnn.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Huấn luyện mô hình
history_cnn = model_cnn.fit(X_train_cnn, y_train, epochs=20, batch_size=32, validation_data=(X_test_cnn, y_test))
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ loss_train và loss_val
plt.plot(history_cnn.history['loss'], label='Loss (Training Data)')
plt.plot(history_cnn.history['val_loss'], label='Loss (Validation Data)')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### RNN model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout

# Xây dựng mô hình RNN
model_rnn = Sequential()

# Layer 1: Tầng SimpleRNN
model_rnn.add(SimpleRNN(64, activation='relu', return_sequences=True, input_shape=(X_train_cnn.shape[1], 1)))

# Layer 2: Tầng Dropout
model_rnn.add(Dropout(0.3))

# Layer 3: Tầng SimpleRNN
model_rnn.add(SimpleRNN(64, activation='relu', return_sequences=True))

# Layer 4: Tầng Dropout
model_rnn.add(Dropout(0.4))

# Layer 5: Tầng SimpleRNN
model_rnn.add(SimpleRNN(32, activation='relu'))

# Layer 6: Tầng Dropout
model_rnn.add(Dropout(0.4))

# Layer 7: Tầng Dense (fully connected)
model_rnn.add(Dense(1))  # Dự đoán giá nhà (giá trị liên tục)

# Biên dịch mô hình
model_rnn.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Huấn luyện mô hình
history_rnn = model_rnn.fit(X_train_cnn, y_train, epochs=20, batch_size=32, validation_data=(X_test_cnn, y_test))
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ loss_train và loss_val
plt.plot(history_rnn.history['loss'], label='Loss (Training Data)')
plt.plot(history_rnn.history['val_loss'], label='Loss (Validation Data)')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### LSTM model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Xây dựng mô hình LSTM
model_lstm = Sequential()

# Layer 1: Tầng LSTM
model_lstm.add(LSTM(64, activation='relu', return_sequences=True, input_shape=(X_train_cnn.shape[1], 1)))

# Layer 2: Tầng Dropout
model_lstm.add(Dropout(0.2))  # Tăng dropout để giảm overfitting

# Layer 3: Tầng LSTM
model_lstm.add(LSTM(32, activation='relu', return_sequences=False))

# Layer 4: Tầng Dropout
model_lstm.add(Dropout(0.3))  # Dropout cao hơn cho tầng cuối

# Layer 5: Tầng Dense (fully connected)
model_lstm.add(Dense(1))  # Đầu ra dự đoán giá trị liên tục

# Biên dịch mô hình
optimizer = Adam(learning_rate=0.0003)  # Giảm learning rate để hội tụ ổn định hơn
model_lstm.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# Callback
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Huấn luyện mô hình
history_lstm = model_lstm.fit(
    X_train_cnn, y_train,
    epochs=40,
    batch_size=32,
    validation_data=(X_test_cnn, y_test),
    callbacks=[early_stopping]
)
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ loss_train và loss_val
plt.plot(history_lstm.history['loss'], label='Loss (Training Data)')
plt.plot(history_lstm.history['val_loss'], label='Loss (Validation Data)')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
import pickle

# Load các mô hình đã lưu
# Load các mô hình đã lưu
model_cnn_load = load_model('model_cnn_housing_hadongv5.keras')
model_rnn_load = load_model('model_rnn_housing_hadongv5.keras') 
model_lstm_load = load_model('model_lstm_housing_hadongv5.keras')
print(model_cnn_load.input_shape)
print(model_rnn_load.input_shape)
print(model_lstm_load.input_shape)


# Định nghĩa các features cần scale
to_be_scaled = ['num_floors', 'num_bed_rooms', 'squared_meter_area', 
                'length_meter', 'width_meter']

#Load scalers đã được fit
with open('predictor_scalerv5.pkl', 'rb') as f:
    PredictorScaler = pickle.load(f)
    
with open('target_scalerv5.pkl', 'rb') as f:
    TargetVarScaler = pickle.load(f)

def predict_house_price(house_data, model, PredictorScaler, TargetVarScaler, to_be_scaled):
    if isinstance(house_data, dict):
        input_data = pd.DataFrame([house_data])
    
    # Tạo DataFrame với các features số học
    numeric_data = pd.DataFrame()
    numeric_data[to_be_scaled] = input_data[to_be_scaled]
    
    # Tạo các cột one-hot cho housing_type
    housing_types = ['Nhà biệt thự', 'Nhà mặt phố', 'Nhà ngõ, hẻm', 'Nhà mặt phố, mặt tiền']
    for housing_type in housing_types:
        col_name = f'housing_type_{housing_type}'
        numeric_data[col_name] = (input_data['housing_type'] == housing_type).astype(int)
    
    # Tạo các cột one-hot cho legal_paper
    legal_papers = ['Giấy tờ khác', 'Đang chờ sổ', 'Đã có sổ']
    for legal_paper in legal_papers:
        col_name = f'legal_paper_{legal_paper}'
        numeric_data[col_name] = (input_data['legal_paper'] == legal_paper).astype(int)
    
    # Tạo các cột one-hot cho ward
    wards = ['Biên Giang', 'Dương Nội', 'Hà Cầu', 'Kiến Hưng', 'La Khê', 
             'Mỗ Lao', 'Nguyễn Trãi', 'Phú La', 'Phú Lãm', 'Phú Lương', 
             'Phúc La', 'Quang Trung', 'Vạn Quán', 'Văn Phúc', 'Yên Nghĩa',
             'Yết Kiêu', 'Đồng Mai']
    
    for ward in wards:
        col_name = f'ward_{ward}'
        numeric_data[col_name] = (input_data['ward'] == ward).astype(int)
    
    # Thêm một cột padding để đạt đủ 30 features
    numeric_data['padding'] = 0
    
    # Scale các features số học
    numeric_data[to_be_scaled] = PredictorScaler.transform(numeric_data[to_be_scaled])
    
    # Chuyển thành numpy array
    input_array = numeric_data.values.astype("float32")
    
    # Reshape cho model (batch_size, timesteps, features)
    input_model = input_array.reshape((1, 30, 1))
    
    # Dự đoán
    predicted_scaled = model.predict(input_model)
    
    # Chuyển đổi về giá trị thực
    predicted_price = TargetVarScaler.inverse_transform(predicted_scaled)[0][0]
    
    return predicted_price

# Test với sample house
sample_house = {
    'num_floors': 3,
    'num_bed_rooms': 5,
    'squared_meter_area': 80.0,
    'length_meter': 8.0,
    'width_meter': 10.0,
    'housing_type': 'Nhà mặt phố, mặt tiền',
    'ward': 'La Khê',
    'legal_paper': 'Đã có sổ'
}

# Dự đoán với các mô hình
price_cnn = predict_house_price(sample_house, model_cnn_load, PredictorScaler, TargetVarScaler, to_be_scaled)
price_rnn = predict_house_price(sample_house, model_rnn_load, PredictorScaler, TargetVarScaler, to_be_scaled)
price_lstm = predict_house_price(sample_house, model_lstm_load, PredictorScaler, TargetVarScaler, to_be_scaled)

print(f"Giá dự đoán (CNN): {price_cnn:,.2f} triệu đồng/m2")
print(f"Giá dự đoán (RNN): {price_rnn:,.2f} triệu đồng/m2")
print(f"Giá dự đoán (LSTM): {price_lstm:,.2f} triệu đồng/m2")
```
